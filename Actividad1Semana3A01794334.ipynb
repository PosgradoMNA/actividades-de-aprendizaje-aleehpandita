**Fundamentos de bases de datos y para ciencia de datos.**

Una base de datos es un programa que nos ayuda a almacenar informacion y que esta información se encuentra almacenada de cierto modo. Se compone de tablas de datos que esta onrganizadas por columnas y filas. Para que la ciencia de datos sea posible es necesario que exista el acceso a los datos es por eso la importancia de las bases datos, en donde se puede almacenar toda la información que se encuentra relacionada y de una forma ordenada y controlada. 

**Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

Es una base de datos de gran tamaño que contiene información de diferentes fuentes y toda esta información se une en una sola base de datos para ser utilizada, en esta base de datos se mantiene una arquitectura donde se almacenan estos datos y debido a dicha arquitectura es posible que se nos permita comprender y utilizar estos datos para tomar decisiones o analizar información. La funcion principal de este tipo de base de datos es que sea utilizada para el analisis de datos, ya que en ella se almacena mucha informacion de diferentes fuentes, toda esta información reunida en un mismo lugar para ser analizada y consultada.




```python
input = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'

import pandas as pd
import numpy as np

df = pd.read_csv(input)
df
```

```python
df.isnull().values.any()
df.isnull().any()
```

```python
ndf = df.copy()
ndf
```

Solo la columna X1 tiene todos los datos validos: 

```python
df.isnull().any()
```

#### Desde el x2 hasta el x23 y la Y, de las columnas falta información, guardare los nombres de las columnas y quitare el X1 que es el unico que tiene toda la información y de ahi sacare la mediana de cada columna y entonces asi podre reemplazar con fillna los valores nulos

```python
columns_names = df.columns.values
columns_names = np.array(columns_names)
columns_names = np.delete(columns_names, [0,1], axis=None)
columns_names
```

```python
for k in columns_names:
  wm = ndf[k].mean()
  ndf[k].fillna(value = wm, inplace = True)

ndf
```

Vemos que ya no hay datos nulos:

```python
ndf.isnull().any()
```
