```python
import pandas as pd
import numpy as np

credit_df = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'

df = pd.read_csv(credit_df,index_col = 0)
df.index.name = None
df
```

```python
df_num = df.iloc[:,3:]
df_num.head()
```

Ahora buscamos la correlaci√≥n de las variables:

```python
df_num.corr()
```

Para analizar mejor los datos:

```python
import seaborn as sns
corrs = df_num.corr()
sns.set(rc = {'figure.figsize':(15,10)})
sns.heatmap(corrs, vmin = -1, vmax = 1, cmap = "BuGn", annot= True)
```

Las casillas con verde oscuro o de color blanco son las que est√°n
mayor correlacionadas. Mientras m√°s oscuro es el verde es m√°s estrecha es la correlaci√≥n positiva.

```python
cols = ["X12", "X13", "X14"]
summary = { 
    "Varianzas": df_num[cols].var()[0:3],
    "Valor Min": df_num[cols].min()[0:3],
    "Valor Max": df_num[cols].max()[0:3]
    }
pd.DataFrame(summary, index = cols).transpose()
```

Buscamos la varianza total de todas las variables y voy a calcular la varianza para cada variable que estamos usando y al final sacamos el % que representa la varianza de cada variable respecto al total

```python

t = df_num.var().sum()

var_x12 = df_num.X12.var()
var_x13 = df_num.X13.var()
var_x14 = df_num.X14.var()

print('Varianza Total \t\t:',t.round(2) )
print('Varianza X12 \t:', ((var_x12/t)*100).round(2),'%' )
print('Varianza X13 \t\t:', ((var_x13/t)*100).round(2),'%' )
print('Varianza X14 \t:', ((var_x14/t)*100).round(2),'%' )
```

A continuaci√≥n voy a escalar los fatos para que todos esten en la misma escala, imprimire los primeros 3 paea no imprimir todos los valores

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(df_num)

scaled[:3]
```

```python
scaled_df = pd.DataFrame(scaled, columns=df_num.columns)
scaled_df.head()
```

Confirmar√© si estan en la misma escala repitiendo los pasos anteriores

```python
cols = ["X12", "X13", "X14"]
summary = { 
    "Varianzas": df_num[cols].var()[0:3],
    "Valor Min": df_num[cols].min()[0:3],
    "Valor Max": df_num[cols].max()[0:3]
    }
pd.DataFrame(summary, index = cols).transpose()
```

Gr√°fica, vemos que los puntos que estan m√°s alejados de la l√≠nea de referencia es un valor at√≠pico.

```python
import seaborn as sns
p = sns.regplot(data = df,  
                x = df.X12, 
                y = df.X13
               ) 
p
```

Ahora con los datos ya en la misma escala, vamos a aplicar PCA, limpiando los valores vacios para que no nos de error

```python
from sklearn.decomposition import PCA
pcs = PCA()

pcs_t = pcs.fit_transform(scaled_df)
pcs_t[:3]
```

```python
pcsSummary_df = pd.DataFrame({'% varianza explicada': np.round(pcs.explained_variance_ratio_,4) * 100,'% varianza acumulada': np.cumsum(pcs.explained_variance_ratio_) * 100})
pcsSummary_df
```

Voy agregar etiquetas a cada fila para poder identificar de una mejor manera

```python
pcs_labels = [f'PC{i + 1}' 
              for i in range(len(scaled_df.columns))]
pcsSummary_df.index = pcs_labels
pcsSummary_df
```

PCA lo que hace es crear componentes de informaci√≥n provenientes de las diferentes variables de nuestro conjunto de datos (por eso se llaman componentes, porque se componen de las varianzas de las variables originales).
Esto quiere decir que los componentes PC1, PC2‚Ä¶PC13 NO SON LAS VARIABLES.

Dataframe de cada componente y vemos que no se perdi√≥ informaci√≥n:

```python
pcs_df = pd.DataFrame(pcs_t, columns =pcs_labels)
print("Varianza total variables originales: ", scaled_df.var().sum())
print("Varianza total de los componentes: ", pcs_df.var().sum())
```

Determino cu√°les son los componentes principales, seleccionamos los componentes que expliquen el 80% o m√°s de la variaci√≥n del conjunto de datos. Esto lo podemos hacer viendo la tabla de las varianzas acumuladas:

```python
pcsSummary_df
```

Vamos que en los primero 8 alrededor del 80% de la variaci√≥n se ve explicada comparaci√≥n con las variables originales pero escaladas, para ver cuantas necesitar√≠amos utilizar para explicar cerca del 80% de la variaci√≥n

```python
total_var =scaled_df.var().sum()
pd.DataFrame({
"Porcentaje Varianza": (scaled_df.var()/ total_var) * 100,
"Porcentaje Varianza Acumulado": (scaled_df.var().cumsum() / total_var) *100
})
```

Logramos reducir casi la mitad de los factores a utilizar para entrenar un modelo, solo nos quedaremos con los primeros 8 componentes

```python
comps_df = pd.DataFrame( pcs.components_.round(4), columns = pcs_df.columns, index = scaled_df.columns)
comps_df.iloc[:,:8]
```
